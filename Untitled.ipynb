```python
from model_data import DataManager
from settings import settings
from word_count import get_meta_info, HighPerformanceCounter
```

```python
# import some handy dandy viz stuff

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
meta_info = get_meta_info()
data_manager = DataManager(settings['experiment_temp_path'],
                          settings['experiment_save_path'])
```

```python
source_by_count = sorted(meta_info.items(), 
                         key=lambda i: sum(i[1]['word_count'].values()),
                         reverse=True)
source_by_count = [(s, sum(info['word_count'].values())) for s, info in source_by_count]
accuracies = [data_manager.get_source_model_accuracy(s)[1] for s, _ in source_by_count]
# sample_structure = {
#     Source.NYT: {
#         'word_count': HighPerformanceCounter('men'=300),
#         datetime.datetime(1993, 8, 12): {
#             'doc_count': 1,
#             'word_count': 300
#         }
#     }
# } 
```

```python
comparative_count = pd.DataFrame({'word_count': [count for s, count in source_by_count]},
                                 index=[s.name for s, _ in source_by_count])
comparative_accuracy = pd.DataFrame({'accuracy': accuracies},
                                    index=[s.name for s, _ in source_by_count])
comparative_count.plot(kind='bar')
comparative_accuracy.plot(kind='bar')
```

```python
ACCUR_CUTOFF_POINT = .65
source_by_accuracy = [(s, a) for (s, _), a in zip(source_by_count, accuracies)]
sources_to_count = []
for source, accuracy in source_by_accuracy:
    if accuracy > ACCUR_CUTOFF_POINT:
        sources_to_count.append(source)

all_dates = set()
for source in sources_to_count:
    all_dates |= set(meta_info[source].keys()) - set(['word_count'])
all_dates = sorted(all_dates)

source_series = dict()
for source in sources_to_count:
    date_counts = {d: 0 for d in all_dates}
    for date in all_dates:
        if date in meta_info[source]:
            date_counts[date] += meta_info[source][date]['word_count']
    date_counts = [date_counts[d] for d in all_dates]
    source_series[source] = date_counts

dates_histo = pd.DataFrame(source_series, index=all_dates)
dates_histo.groupby([dates_histo.index.year])\
           .sum()\
           .plot(kind='bar', figsize=(7,5))
```
