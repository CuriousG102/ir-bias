```python
from analysis import BiasFinder
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
def bias_plot_normalized(neutral_words_path):
    bias_finder = BiasFinder(neutral_words_path=neutral_words_path)
    sources = sorted(bias_finder.get_eligible_sources())
    direct_bias = [bias_finder.get_source_bias_measurement(source) for source in sources]
    normed_direct_bias = [float(i)/sum(direct_bias) for i in direct_bias]
    normed_source_bias = pd.DataFrame({'normed_direct_bias': normed_direct_bias},
                                      index=sources)
    normed_source_bias.plot(kind='bar')
```

```python
bias_plot_normalized('/newsdata/crawford_gend_neutral.csv')
```

```python
bias_plot_normalized('/newsdata/occupations.csv')
```

```python
bias_plot_normalized('/newsdata/pleasant_unpleasant.csv')
```
