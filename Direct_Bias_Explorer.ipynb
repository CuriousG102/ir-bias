```python
from analysis import BiasFinder
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
bias_finder = BiasFinder()
sources = bias_finder.get_eligible_sources
# For Crawford neutral word list
direct_bias = [bias_finder.get_source_bias_measurement(source) for source in sources]
```

```python
source_bias = pd.DataFrame({'direct_bias': direct_bias}, 
                          index = sources)
source_bias.plot(kind='bar')
```

```python
normed_direct_bias = [float(i)/sum(direct_bias) for i in direct_bias]
normed_source_bias = pd.DataFrame({'direct_bias': normed_direct_bias}, 
                          index = sources)
normed_source_bias.plot(kind='bar')
```
