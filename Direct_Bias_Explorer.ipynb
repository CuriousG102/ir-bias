```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
crawf_bias_finder = BiasFinder(neutral_words_path='/newsdata/crawford_gend_neutral.csv')
crawf_sources = bias_finder.get_eligible_sources()
# For Crawford neutral word list
crawf_direct_bias = [crawf_bias_finder.get_source_bias_measurement(crawf_source) for crawf_source in crawf_sources]
```

```python
crawf_source_bias = pd.DataFrame({'direct_bias': crawf_direct_bias}, 
                          index = crawf_sources)
crawf_source_bias.plot(kind='bar')
```

```python
crawf_normed_direct_bias = [float(i)/sum(crawf_direct_bias) for i in crawf_direct_bias]
crawf_normed_source_bias = pd.DataFrame({'direct_bias': crawf_normed_direct_bias}, 
                          index = crawf_sources)
crawf_normed_source_bias.plot(kind='bar')
```
