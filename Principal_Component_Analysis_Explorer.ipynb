```python
from analysis import BiasFinder
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
bias_finder = BiasFinder()
sources = bias_finder.get_eligible_sources()
explained_variance = []
for source in sources:
    pca = bias_finder.get_principal_components(source)
    explained_variance = pd.DataFrame({'explained_variance': pca.explained_variance_ratio_}, 
                                      index = [x + 1 for x in range(len(pca.explained_variance_ratio_))])  
    print(source)
    explained_variance.plot(kind='bar')
    plt.show()
```
