```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
bias_finder = BiasFinder()
sources = bias_finder.get_eligible_sources()
first_components = []
for source in sources:
    model = bias_finder.data_manager.get_model_for_source(source)
    pca = bias_finder.get_principal_components(model)
    explained_variance = pd.DataFrame({'explained_variance': pca.explained_variance_ratio_}, 
                                      index = [x + 1 for x in range(len(pca.explained_variance_ratio_))])  
    print(source)
    explained_variance.plot(kind='bar')
    first_components.append(pca.explained_variance_ratio_[0])
    plt.show()
```

```python
np.mean(first_components)
```
